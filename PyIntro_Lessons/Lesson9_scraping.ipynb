```python
# Requests is a python library, mainly for fetching information from 
# the internet. Requests is very often used. Not only in scraping. 
# (Preview: Often used with APIs)

import requests

# With requests we can use the get-method to fetch information
# from a url we provide
response = requests.get("https://quotes.toscrape.com/")
# The response is a response object with a status code, 200 is good, (404 is bad)
print(response)
```

```python
# with response.content we can find the html from the fetched website
print(response.content)
```

```python
# The response content is painful to read and handle, so we use
# BeautifulSoup to parse the html and make it easier to work with
from bs4 import BeautifulSoup

# BeautifulSoup(html, parser) parses the html with the parser we provide
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())  
```

Below is a cell that illustrates how tags in beautifulsoup work, they create objects.

```python
# BeautifulSoup also helps us create objects that represent HTML tags
# and provides methods and attributes to work with them.

tag = soup.find('h2')  # Find the first <h1> tag
print(tag.name)
tag.name = "fiskpinne"  # Change the tag name from <h1> to <h2>
print("The full tag:", tag)

```

Let's find all the quotes on the page and print them out.
===

```python
# soup.find() finds the first instance of a tag
# soup.find_all() finds all instances of a tag
quotes = soup.find_all("div", class_="quote")  # Find the first <div> tag with class "quote"


# Now that we have a big list of all quotes, we can loop through them
for quote in quotes:
    quote_text = quote.find("span", class_="text").get_text()  # Get the text of the quote
    author = quote.find("small", class_="author").get_text() # Get the author of the quote
    print(f"{quote_text} - {author}")  # Print the quote and author
```

```python
import requests
from bs4 import BeautifulSoup

# The URL we want to scrape
url = 'http://quotes.toscrape.com/'
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print("Successfully fetched the page.")
else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```

```python
print(response.content)
```

```python
# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Find all the quote divs
quotes = soup.find_all('div', class_='quote')

# Loop through each quote to extract the data
for quote in quotes:
    # Find the quote text, which is inside a span with class 'text'
    text = quote.find('span', class_='text').text
    # Find the author, which is inside a small tag with class 'author'
    author = quote.find('small', class_='author').text
    print(f'"{text}" - {author}')
```
