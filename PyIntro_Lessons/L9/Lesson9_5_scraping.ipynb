Make it robust
=================

Not all quotes have tags

```python
import requests
from bs4 import BeautifulSoup

url = 'http://quotes.toscrape.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

quotes = soup.find_all('div', class_='quote')

for quote in quotes:
    # 1. Clean the text and handle potential errors
    text_element = quote.find('span', class_='text')
    if text_element:
        # Use strip to remove leading/trailing whitespace and replace to clean quotes
        text = text_element.text.strip().replace('“', '').replace('”', '')
    else:
        text = "Quote text missing."
        
    author = quote.find('small', class_='author').text
    
    # 2. Handle missing tags gracefully
    tags_element = quote.find('div', class_='tags')
    
    tags = []
    if tags_element:
        # Extract all tag names
        for tag_a in tags_element.find_all('a', class_='tag'):
            tags.append(tag_a.text)
    
    print(f'Author: {author}')
    print(f'Quote: {text}')
    print(f'Tags: {", ".join(tags) if tags else "No Tags"}')
    print('-' * 20)
```

### Activity 3: The "Scraping a List" Challenge

**Objective:** Create a Python script that scrapes the top 50 movies from a simple list, like the one on [**IMDb's Top 250 list**](https://www.imdb.com/chart/top/).

**Instructions:**

  * **Step 1:** Use `requests` to get the page's HTML.
  * **Step 2:** Use BeautifulSoup to parse the HTML.
  * **Step 3:** Use the "Inspect" tool to identify the HTML element that contains each movie title. (Hint: Look for a list item `<li>` or a table row `<tr>`.)
  * **Step 4:** Find the tags and classes that hold the movie **title** and the movie's **rating**.
  * **Step 5:** Loop through the data you find and print the title and rating for each movie.



```python
import requests
from bs4 import BeautifulSoup
import json

url = 'https://www.imdb.com/chart/top/'

try:
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')

    # print(soup.prettify()[:1500])  # Check the initial part of the HTML content

    tag = soup.find('title')
    print(tag.prettify())

    # Find the JSON-LD script tag
    script_tag = soup.find('script', type='application/ld+json')
    if script_tag:
        data = json.loads(script_tag.string) # New and weird, uses json library
        movies = data.get('itemListElement', [])
        counter = 0
        for movie in movies[:50]:
            item = movie.get('item', {})
            title = item.get('name', 'N/A')
            rating = item.get('aggregateRating', {}).get('ratingValue', 'N/A')
            counter += 1
            print(f"Rank {counter} Title: {title}, Rating: {rating}")
    else:
        print("Could not find JSON-LD data.")

except requests.exceptions.RequestException as e:
    print(f"Error fetching the URL: {e}")
```

Pagination
=========
Finding data on multiple pages

Assignment: Scrape all quotes from all pages on [Quotes to Scrape](http://quotes.toscrape.com/)

Hint: You can use a while loop maybe?

```python
import requests
from bs4 import BeautifulSoup

# Url for page five is https://quotes.toscrape.com/page/5/

base_url = 'http://quotes.toscrape.com'
all_quotes = []

# Scrape the first 3 pages
for page_number in range(1, 4):
    url = f'{base_url}/page/{page_number}/'
    print(f"--- Scraping {url} ---")
    
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to fetch page {page_number}. Stopping.")
        break

    soup = BeautifulSoup(response.text, 'html.parser')
    
    quotes = soup.find_all('div', class_='quote')
    
    for quote in quotes:
        # Simple extraction for demonstration
        text = quote.find('span', class_='text').text
        author = quote.find('small', class_='author').text
        all_quotes.append({'text': text, 'author': author, 'page': page_number})

# Print a summary of the collected data
print(f"\nSuccessfully collected {len(all_quotes)} quotes across 3 pages.")
```

```python
# Print last 5 quotes as a sample
for quote in all_quotes[-5:]:
    print(f"Page {quote['page']}: {quote['text']} - {quote['author']}")

# Print all quotes for each page
for page in range(1, 4):
    print(f"\nQuotes from page {page}:")
    for quote in all_quotes:
        if quote['page'] == page:
            print(f"- {quote['text']} by {quote['author']}")
```

Creating an Class for Scraping
==============================

```python
import requests
from bs4 import BeautifulSoup

class QuoteScraper:
    """A class designed to scrape quotes from quotes.toscrape.com."""
    
    def __init__(self, base_url):
        self.base_url = base_url
        self.all_data = []

    def _fetch_page(self, url):
        """Private method to fetch and parse HTML."""
        try:
            response = requests.get(url)
            response.raise_for_status() # Check for bad status codes
            return BeautifulSoup(response.text, 'html.parser')
        except requests.exceptions.RequestException as e:
            print(f"Error fetching {url}: {e}")
            return None

    def scrape_page(self, page_num):
        """Scrapes data from a single, numbered page."""
        url = f'{self.base_url}/page/{page_num}/'
        soup = self._fetch_page(url)
        
        if not soup:
            return [] # Return empty list on error

        # The below code you could modify to suit your needs
        page_quotes = []
        for quote in soup.find_all('div', class_='quote'):
            text = quote.find('span', class_='text').text.strip().replace('“', '').replace('”', '')
            author = quote.find('small', class_='author').text
            page_quotes.append({'author': author, 'quote': text, 'page': page_num})
            
        return page_quotes

    def run_scraper(self, max_pages=5):
        """Runs the scraper for a specified number of pages."""
        print(f"Starting scrape for up to {max_pages} pages...")
        for page in range(1, max_pages + 1):
            data = self.scrape_page(page)
            if not data:
                print(f"No data found on page {page}. Stopping.")
                break # Stop if a page fails or is empty
            self.all_data.extend(data)
            print(f"Collected {len(data)} quotes from page {page}.")
        print(f"Scraping finished. Total quotes collected: {len(self.all_data)}")


# EXECUTION: The main script becomes clean and readable!
if __name__ == '__main__':
    scraper = QuoteScraper('http://quotes.toscrape.com')
    scraper.run_scraper(max_pages=2) # Only scrape the first two pages
    
    # Example: print the data
    # for item in scraper.all_data:
    #     print(f"[{item['page']}] {item['author']}: {item['quote'][:50]}...")
```
